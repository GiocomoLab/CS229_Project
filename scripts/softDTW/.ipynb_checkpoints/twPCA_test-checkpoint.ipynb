```python
from twpca import TWPCA
import scipy.io as sio
import numpy as np
from scipy.ndimage.filters import gaussian_filter1d
from twpca.datasets import jittered_neuron
from sdtw.distance import SquaredEuclidean

%pylab inline
%load_ext autoreload
%autoreload 2
```

```python
baseDir = "/Users/markplitt/Dropbox/Malcolms_VR_data/twPCA_Mats/"
params = sio.loadmat(baseDir+"params.mat")
params.keys()
for i, arr in enumerate(params['grid_fnames']):
    params['grid_fnames'][i]= params['grid_fnames'][i][0][0]
    
    
```

```python
# load all grid cell data in dictionary 
spikeMats, frMats = [],[]
for i, fname in enumerate(params['grid_fnames']):
    matDat = sio.loadmat(baseDir+fname[0])
    spikeMats.append(gaussian_filter1d(matDat['spikeMat'],10,axis = 1))
    frMats.append(matDat['posMat'][0])
    

# plot distribution of number of trials
numTrials = [spikeMats[i].shape[0] for i in range(len(spikeMats))]
maxLengths = [spikeMats[i].shape[1] for i in range(len(spikeMats))]
print(np.min(numTrials))
print(np.max(numTrials))
print(np.min(maxLengths))
print(np.max(maxLengths))
hist(numTrials)
```

```python
data={}
data['aligned_dat']= []
data['time_factors'] = []
data['fnames'] = params['grid_fnames']
data['spike_rate_raw'] = spikeMats
data['posMat_raw'] = posMats
for i in range(10):
    raw_data = np.expand_dims(spikeMats[i],axis=2)
#raw_data[np.isnan(raw_data)] = 0.
#imshow(np.squeeze(raw_data))

    n_components = 1
    model = TWPCA(raw_data,1).fit()

    
    data['time_factors'].append(model.params['time'])         
    

    # you can use the model object to align data (compare this to the aligned_data from above)
    estimated_aligned_data = model.transform()
    data['aligned_dat'].append(np.squeeze(estimated_aligned_data))

    
```
